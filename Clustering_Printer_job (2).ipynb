```python
job_names = df_print['JobName'].dropna().astype(str).tolist()
```

```python
embeddings = np.load("job_embeddings.npy")
```

```python
from sentence_transformers import SentenceTransformer

# Load the embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate sentence embeddings
embeddings = model.encode(job_names, show_progress_bar=True)
```

```python
df_embeddings = pd.DataFrame(embeddings)
```

```python
df_embeddings.shape
```

```python
import numpy as np

# Save
np.save("job_embeddings.npy", embeddings)

# Load later
embeddings = np.load("job_embeddings.npy")
```

```python
import hdbscan

# Create and fit the HDBSCAN model
clusterer = hdbscan.HDBSCAN(min_cluster_size=30, metric='euclidean')
labels = clusterer.fit_predict(embeddings)

# Add cluster labels to your DataFrame
df_print['cluster'] = labels
```
